```python
from google import genai
from google.genai import types
import requests
import json
```

```python
client = genai.Client(api_key='YOUR_API_KEY')
chat = client.chats.create(model="gemini-2.0-flash", 
                        config=types.GenerateContentConfig(
                               system_instruction=
                               """
                               Ты будешь получать описание работы в сфере IT. Твоя задача выделить ключевые навыки, необходимые для данной вакансии.
                               Возвращай только список навыков, разделенных запятой без пробелов (как для csv файла). Старайся называть одинаковые навыки одинакого.
                               Например, если в одной вакансии встретился навык Linux, а в другой Линукс, называй обоих Linux.
                               Выедляй до 10 самых важных навыков для каждой вакансии. 
                                """,
                               temperature=1.0)
                            )
```

```python
header = {'HH-User-Agent': 'SkillsAnalysis/1.0 (DoIReallyNeedThis@yandex.ru)'}
response = requests.get('https://api.hh.ru/vacancies/120046927', headers=header)
```

```python
skills = chat.send_message(json.loads(response.text)['description'])
print(skills.text)
```

```python
ids = []
for i in range(10):
    try:
        a = requests.get('https://api.hh.ru/vacancies', params={
            'professional_role': '116', # 116 for cyber security, 165 for data scientists
            'page': i,
            'per_page': 100,
            'area': ["113", "40", "16"] # Russia, Belarus, Kazakhstan
        },
        headers=header)
        vacancies_data = json.loads(a.text)['items']

        for f in vacancies_data:
            ids.append(f['id'])
    except Exception as e:
        print(e)
        break
print('Number of jobs: ', len(ids))
```

```python
response = requests.get('https://api.hh.ru/vacancies/' + str(ids[1]), headers=header)
print(json.dumps(json.loads(response.text), indent=4, ensure_ascii=False))
```

```python
cnt = 0
with open('skills.csv', 'w', encoding='utf-8') as f_skills, open('salary.csv', 'w', encoding='utf-8') as f_salary:
    for id in ids:
        cnt = cnt + 1
        print('{}/{}'.format(cnt, len(ids)))
        response = requests.get('https://api.hh.ru/vacancies/' + str(id), headers=header)
        response_json = json.loads(response.text)
        skills = chat.send_message(response_json['description'])
        f_skills.write(skills.text)
        if ('salary' in response_json):
            f_salary.write(str(response_json['salary']))
            f_salary.write('\n')
        else:
            f_salary.write('None\n')
```

```python
client2 = genai.Client(api_key='YOUR_API_KEY')
chat2 = client2.chats.create(model="gemini-2.0-flash", 
                        config=types.GenerateContentConfig(
                               system_instruction=
                               """
                               Ты будешь получать на вход различные навыки, важные для дата сайнтиста. Некоторые из них будут означать одно и то же, но
                               иметь разные названия. Если тебе уже поступал такой навык, но под другим названием, то верни изначально название и ничего больше.
                               Иначе верни название, поступившее на вход. 
                                """)
                            )

cnt = 0
with open('skills.csv', 'r') as dirty, open('skills_cleaned.csv', 'w') as clean:
    for skills in dirty:
        print(cnt)
        cnt += 1
        a = list(skills.split(',')[:-1])
        for skill in a:
            clean.write(chat2.send_message(skill).text[:-1] + ',')
        clean.write('\n')
```
